```python
#a*b
def multiply(a,b):

  result = 0

  for i in range(0,b):
    result = result + a
  return result
multiply(3,4)
```

```python
#recursion

def mul(a,b):
  if b == 1:
    return a
  else:
    return a + mul(a,b-1)
print(mul(5,6))
```

```python
#factorial

def fact(number):
  if number == 1:
    return 1
  else:
    return number * fact(number - 1)
print(fact(4))
```

```python
def fact(num):

  if num == 1:
    return 1
  else:
    return num * fact(num-1)
fact(5)
```

```python
#Palindrome
def palin(text):
  if len(text) < 1:
    print('palindrome')
  else:
    if text[0] == text[-1]:
      palin(text[1:-1]) #removing tested text
    else:
      print('not a palindrome')
palin('malayalam')
palin('abba')
```

```python
#fibonacci

def fib(m):
  if m == 0 or m ==1:
    return 1
  else:
    return fib(m-1) + fib(m-2)
fib(5)
```

```python
def memo(m,d):
  if m in d:
    return d[m]
  else:
    d[m] = memo(m-1,d) + memo(m-2,d)
    return d[m]

d = {0:1,1:1}

memo(48,d)
```

```python
def power_set(lst):
    if len(lst) == 0:
        return [[]]

    first = lst[0]
    rest = power_set(lst[1:])

    return rest + [[first] + subset for subset in rest]

L = [1, 2, 3]
print(power_set(L))
```
