### Exceptions

If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.

- Exceptions are raised by python runtime
- You have to takle is on the fly

#### **Examples**

- Memory overflow
- Divide by 0 -> logical error
- Database error

```python
# Why is it important to handle exceptions
# how to handle exceptions
# -> Try except block
```

```python
# let's create a file
with open('sample.txt','w') as f:
  f.write('hello world')
```

```python
# try catch demo
try:
  with open('sample1.txt','r') as f:
    print(f.read())
except:
  print('sorry file not found')
```

```python
# catching specific exception -> different types of errors ke liye different except-> sarkari coder -> one error message only

try:
  m=5
  f = open('sample.txt','r')
  print(f.read())
  print(m)
  print(5/2)
  L = [1,2,3]
  L[100]
except FileNotFoundError: #agr ye error aaya
  print('file not found') #to ye print hoga
except NameError:
  print('variable not defined')
except ZeroDivisionError:
  print("can't divide by 0")

except Exception as e: #generic exception last me hon achahiye' #ye bata dega ki kaisa error hai agr predefined nhi hai to
  print(e)
  print(e.with_traceback)
```

Note: Every error is an class

```python
class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise Exception('amount cannot be -ve') #a class #throwing the error
    if self.balance < amount:
      raise Exception('paise nai hai tere paas')
    self.balance = self.balance - amount

obj = Bank(10000)


try:
  obj.withdraw(-500) #i know yha error ho skta hai isliye maine try except me daala
except Exception as e: #catching the error beign throwed #e is the object of the Exception class
  print(e)
else:
  print(obj.balance)
```

custom Exception Class

```python
class MyException(Exception): #have to inherit from Exception class

  def __init__(self,message):
    print(message)

class Bank:

  def __init__(self,balance):
    self.balance = balance

  def withdraw(self,amount):
    if amount < 0:
      raise MyException('amount cannot be -ve') #here making the object of MyException Class jo message print kr rha hai
    if self.balance < amount:
      raise MyException('paise nai hai tere paas')
    self.balance = self.balance - amount

obj = Bank(10000)
try:
  obj.withdraw(-15000)
except MyException as e:
  pass
else:
  print(obj.balance)
```

Why create Custom Exception Class? - We create custom exception classes according to our application need- > if apart from security error we also have to provide other functionalities

```python
class SecurityError(Exception):
   #you can add methods inside Exception Class
  def __init__(self,message):
    print(message)

  def logout(self):
    print('logout')


class Google:

  def __init__(self,name,email,password,device):
    self.name = name
    self.email = email
    self.password = password
    self.device = device

  def login(self,email,password,device):
    if device != self.device:
      raise SecurityError('bhai teri to lg gyi')
    if email == self.email  and password == self.password:
      print('welcome')
    else:
      print('login error')


obj = Google('nitish','nitish@gmail.com','1234','android')

try:
  obj.login('nitish@gmail.com','1234','windows')
except SecurityError as e: #e is the object of SecurityError
  e.logout()
else:
  print(obj.name)
finally:
  print('database connection closed')
```
