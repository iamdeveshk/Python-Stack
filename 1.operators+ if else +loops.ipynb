```python
# guessing game between 1 and 100
import random

jackpot = random.randint(1,100)
guess= int(input("Enter the guess number:"))

#loop for game
counter = 1
while guess != jackpot:
    if guess < jackpot:
        print('Galat!!, guess higher')
    else:
        print('Galat!!, guess lower')
    guess = int(input("Enter the guess number:"))
    counter +=1
else:
    print('Correct Guess', jackpot)
    if counter >= 5:
        print('Visuddh Vhutiya ho be itne baar me guess kiye: ',counter)
    else:
        print('Number of attempts:',counter)
```

```python

rows=int(input("Enter number of rows:"))
for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end='')
    for k in range(i-1,0,-1):
        print(k,end='')
    print()

```

```python


rows=int(input("Enter number of rows:"))
for i in range(1,rows+1): #outer loop controls number of rows (how many times the loop will run)
    for j in range(1,i+1): #the inner loop controls the number of stars printed
        print('*',end='')
    print() #this will print next line as there is nothing to print ##changing line
```

```python
#program to find prime numbers between a given range

lower=int(input("Enter lower range:"))
upper=int(input("Enter upper range:"))

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i)
    
```

```python
#continue
for i in range(1,10):
    if i==5:  #skips the iteration at i == 5
        continue
    print(i)
```
